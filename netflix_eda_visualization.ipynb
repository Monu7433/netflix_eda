```python
# ==============================
# IMPORT LIBRARIES
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
# ==============================
# 1. Count of Movies vs TV Shows
# ==============================

# Matplotlib
counts = df["type"].value_counts()

plt.figure()
plt.bar(counts.index, counts.values)
plt.xlabel("Type")
plt.ylabel("Count")
plt.title("Count of Movies vs TV Shows (Matplotlib)")
plt.show()


# Seaborn
plt.figure()
sns.countplot(data=df, x="type")
plt.title("Count of Movies vs TV Shows (Seaborn)")
plt.show()
```

```python
# ==============================
# 2. Distribution of Release Year
# ==============================

# Matplotlib
plt.figure()
plt.hist(df["release_year"], bins=20)
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.title("Distribution of Release Year (Matplotlib)")
plt.show()


# Seaborn
plt.figure()
sns.histplot(data=df, x="release_year", bins=20)
plt.title("Distribution of Release Year (Seaborn)")
plt.show()
```

```python
# ==============================
# 3. Boxplot: Release Year by Type
# ==============================

# Matplotlib
movie_years = df[df["type"] == "Movie"]["release_year"]
tv_years = df[df["type"] == "TV Show"]["release_year"]

plt.figure()
plt.boxplot([movie_years, tv_years], labels=["Movie", "TV Show"])
plt.ylabel("Release Year")
plt.title("Release Year by Type (Matplotlib)")
plt.show()


# Seaborn
plt.figure()
sns.boxplot(data=df, x="type", y="release_year")
plt.title("Release Year by Type (Seaborn)")
plt.show()
```

```python
# ==============================
# 4. Top 10 Countries
# ==============================

top_countries = df["country"].value_counts().head(10)

# Matplotlib
plt.figure()
plt.barh(top_countries.index, top_countries.values)
plt.xlabel("Count")
plt.title("Top 10 Countries (Matplotlib)")
plt.show()


# Seaborn
plt.figure()
sns.countplot(
    data=df[df["country"].isin(top_countries.index)],
    y="country",
    order=top_countries.index
)
plt.title("Top 10 Countries (Seaborn)")
plt.show()
```

```python
# ==============================
# 5. Correlation Heatmap
# ==============================

numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()

# Matplotlib
plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix (Matplotlib)")
plt.show()


# Seaborn
plt.figure()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Seaborn)")
plt.show()

```
